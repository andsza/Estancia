```python
lista_acc=[] 
lista_desc=[]
lista_age=[]
lista_sex=[]
lista_bmi=[]
lista_geo=[]
lista_date=[]
```

```python
for x in lista_ebi:
    with open(ruta_ebi+x,"r") as file: 
        archivo=json.load(file)
        acc=archivo["accession"]
        lista_acc.append(acc)
        desc=archivo["characteristics"]["description"][0]["text"]
        lista_desc.append(desc)
        age=archivo["characteristics"]["host age"][0]["text"]
        lista_age.append(age)
        sex=archivo["characteristics"]["host sex"][0]["text"]
        lista_sex.append(sex)
        bmi=archivo["characteristics"]["bmi"][0]["text"]
        lista_bmi.append(bmi)
        geo=archivo["characteristics"]["geographic location (country and/or sea)"][0]["text"]
        lista_geo.append(geo)
        date=archivo["characteristics"]["collection date"][0]["text"]
        lista_date.append(date)
```
